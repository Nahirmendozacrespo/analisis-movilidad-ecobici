# **Tabla usuarios 2015-2025**

```python
import pandas as pd
import glob

# 1. Definir la ruta específica para buscar solo los archivos de usuarios
# El asterisco (*) atrapará cualquier año en el nombre del archivo
# Asegúrate de que la ruta incluya tu carpeta 'ecobici' y el comodín '*.csv' al final
ruta_usuarios = '/content/drive/MyDrive/ecobici/usuarios-ecobici-*.csv'
archivos_usuarios = glob.glob(ruta_usuarios)
print(f"Se encontraron {len(archivos_usuarios)} archivos de usuarios para unir.")

lista_usuarios = []

for archivo in archivos_usuarios:
    df_temp = pd.read_csv(archivo)

    # 2. Estandarizar el nombre de la columna (por el problema de ID_usuario vs id_usuario)
    if 'ID_usuario' in df_temp.columns:
        df_temp = df_temp.rename(columns={'ID_usuario': 'id_usuario'})

    # 3. Forzar el ID a texto y limpiarlo de decimales (.0) o espacios
    if 'id_usuario' in df_temp.columns:
        df_temp['id_usuario'] = df_temp['id_usuario'].astype(str)
        df_temp['id_usuario'] = df_temp['id_usuario'].str.replace('.0', '', regex=False).str.strip()

    lista_usuarios.append(df_temp)

# 4. Concatenar toda la historia en una sola tabla gigante
df_usuarios_historico = pd.concat(lista_usuarios, ignore_index=True)
total_previo = len(df_usuarios_historico)

# 5. LA PRUEBA DE FUEGO: Eliminar usuarios duplicados
# subset=['id_usuario'] le dice a Pandas que busque repetidos basándose solo en el ID
# keep='last' asegura que si hay duplicados, nos quedamos con la versión más reciente de sus datos
df_usuarios_historico = df_usuarios_historico.drop_duplicates(subset=['id_usuario'], keep='last')
total_final = len(df_usuarios_historico)

print("\n--- RESUMEN DEL PADRÓN MAESTRO ---")
print(f"Registros brutos concatenados: {total_previo:,}")
print(f"Duplicados eliminados: {total_previo - total_final:,}")
print(f"🚴 Usuarios ÚNICOS finales: {total_final:,}")
```

```python
# 2. Convertir ambas a texto (nuestra regla de oro)
df_usuarios_historico['id_usuario'] = df_usuarios_historico['id_usuario'].astype(str)
df_historico2023['id_usuario'] = df_historico2023['id_usuario'].astype(str)
```

```python
# 3. Limpiar la "basura" de la tabla de recorridos
df_historico2023['id_usuario'] = df_historico2023['id_usuario'].str.replace('BAEcobici', '', regex=False)
```

```python
# 1. El Momento de la Verdad: ¿Qué porcentaje de viajes cruza con el Padrón Maestro?
# Usamos isin() para buscar cada ID de recorrido dentro del historial completo de usuarios
viajes_con_match_historico = df_historico2023['id_usuario'].isin(df_usuarios_historico['id_usuario'])

# 2. Calcular el porcentaje de éxito
porcentaje_exito_historico = (viajes_con_match_historico.mean() * 100)

# 3. Mostrar los resultados triunfales
print("--- PRUEBA DE FUEGO CON EL PADRÓN MAESTRO ---")
print(f"Total de viajes en 2023: {len(df_historico2023):,}")
print(f"Viajes que SÍ encontraron a su usuario: {viajes_con_match_historico.sum():,}")
print(f"Tasa de éxito del cruce (JOIN) actualizada: {porcentaje_exito_historico:.2f}%")
```

Tenemos llave para unir las tablas en SQL con ID usuario entre tablas recorridos y usuarios. Tambien tenemos Padron historico del 2015-2025.

El cruce del padron con la tabla del 2023 "id_usuario" dio un 94.34% de confiabilidad.

Limpieza de los dataset recorridos del 2025-2018

```python
recorridos2025 = '/content/drive/MyDrive/ecobici/recorridos/recorridos-realizados-2025.csv'
df_recorridos2025 = pd.read_csv(recorridos2025)
df_recorridos2025.head()
```

```python
df_recorridos2025.info()
```

```python
# 1. Definir la lista de columnas pesadas o innecesarias que vamos a amputar
columnas_a_borrar = [
    'Id_recorrido',
    'long_estacion_origen',
    'lat_estacion_origen',
    'long_estacion_destino',
    'lat_estacion_destino'
]

# Le pedimos a Pandas que las elimine si existen en el dataset
columnas_presentes = [col for col in columnas_a_borrar if col in df_recorridos2025.columns]
df_recorridos2025 = df_recorridos2025.drop(columns=columnas_presentes)

# 2. ELIMINAR LA BASURA DE LOS IDs ("BAEcobici")
columnas_id = ['id_estacion_origen', 'id_estacion_destino', 'id_usuario']
for col in columnas_id:
    if col in df_recorridos2025.columns:
        # Forzamos a texto y reemplazamos la palabra
        df_recorridos2025[col] = df_recorridos2025[col].astype(str).str.replace('BAEcobici', '', regex=False)
        # Limpieza de seguridad por si quedaron decimales fantasma (.0) o espacios
        df_recorridos2025[col] = df_recorridos2025[col].str.replace('.0', '', regex=False).str.strip()

# 3. ESTANDARIZAR NOMBRES (Quitar la tilde de género)
df_recorridos2025 = df_recorridos2025.rename(columns={'género': 'genero'})

# 4. ARREGALAR LA DURACIÓN (De texto 'object' a número)
if 'duracion_recorrido' in df_recorridos2025.columns:
    df_recorridos2025['duracion_recorrido'] = pd.to_numeric(df_recorridos2025['duracion_recorrido'], errors='coerce')

# --- COMPROBACIÓN FINAL ---
print("✅ ¡Cirugía del dataset 2025 completada con éxito!")
print("\n--- NUEVA RADIOGRAFÍA ---")
df_recorridos2025.info()
```

```python
# Convertimos las columnas de fechas de texto plano a formato datetime real
columnas_fecha = ['fecha_origen_recorrido', 'fecha_destino_recorrido']

for col in columnas_fecha:
    # Usamos errors='coerce' para que, si hay alguna fecha escrita con un error de tipeo,
    # la convierta en nulo (NaN) sin frenar el código
    df_recorridos2025[col] = pd.to_datetime(df_recorridos2025[col], errors='coerce')

# Verificamos la transformación
print("--- TIPOS DE DATOS ACTUALIZADOS ---")
print(df_recorridos2025[['fecha_origen_recorrido', 'fecha_destino_recorrido']].dtypes)
```

```python
# 1. Definimos las dos columnas de direcciones que queremos borrar
columnas_direccion = ['direccion_estacion_origen', 'direccion_estacion_destino']

# 2. Le pedimos a Pandas que las elimine
df_recorridos2025 = df_recorridos2025.drop(columns=columnas_direccion)

# 3. Verificamos cómo quedó nuestra base aún más liviana
print("✅ Direcciones eliminadas.")
df_recorridos2025.info()
```

```python
# Definimos la ruta donde se guardará el archivo limpio del 2025
ruta_2025_final = '/content/drive/MyDrive/ecobici/recorridos_2025_LIMPIO.csv'

# Exportamos la tabla sin la columna extra de índices
df_recorridos2025.to_csv(ruta_2025_final, index=False)

print("✅ ¡Dataset de recorridos 2025 exportado y asegurado en Drive!")
```

```python
import glob
```

```python
# 1. Usar glob para buscar todos los archivos YA LIMPIOS de recorridos en tu Drive
# Ajusta el nombre si los guardaste diferente
ruta_limpios = '/content/drive/MyDrive/ecobici/recorridos_limpios/recorridos_*_LIMPIO.csv'
archivos_limpios = glob.glob(ruta_limpios)

print(f"Se encontraron {len(archivos_limpios)} archivos limpios en Drive. Iniciando lectura...\n")

lista_dataframes = []

# 2. Leer cada archivo desde Drive y guardarlo en una lista
for archivo in archivos_limpios:
    print(f"Cargando: {archivo.split('/')[-1]}") # Imprime solo el nombre del archivo
    df_temp = pd.read_csv(archivo, low_memory=False)
    lista_dataframes.append(df_temp)

print("\nTodos los archivos cargados. Iniciando la gran concatenación...")

# 3. La gran concatenación
df_recorridos_historico = pd.concat(lista_dataframes, ignore_index=True)

# 4. Eliminamos la columna modelo_bicicleta de todo el historial de una vez
df_recorridos_historico = df_recorridos_historico.drop(columns=['modelo_bicicleta'], errors='ignore')

# 5. Radiografía final
print("\n✅ ¡Operación exitosa!")
print(f"Total de viajes históricos consolidados: {len(df_recorridos_historico):,}")
print("\n--- RADIOGRAFÍA FINAL ---")
df_recorridos_historico.info()
```
