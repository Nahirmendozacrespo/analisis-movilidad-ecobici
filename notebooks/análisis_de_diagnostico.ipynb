# **¿Por qué no superamos el record historico del 2019?**

La cantidad de viajes sostenidos en el 2025 representan la mitad del volumen de viajes historico del 2019 y se debe a 2 factores.

1. Barrera económica: La trancisión del 100% gratuito a un modelo pago.

2. Déficit de infraestructura intermodal: Faltan conexiones estrategicas entre la red de ciclovias y las entradas claves a la Provincia de Buenos Aires. (Por ejemplo: Conexión con el ferrocaril Belgrano Sur).

Las ecobicis se establecieron dentro de las mismas localidades donde ya tiene posición como Palermo, Caballito, Belgrano, Retiro y Constitución.

Pero no se expandieron a otros puntos estrategicos que son importantes para la vida diaria de miles de trabajadores, estudiantes y personas con fines recreativos. Por ejemplo tenemos Liniers como punto estrategico donde salen muchos colectivos hacia distintos puntos dentro de capital federal y hacia la provincia de buenos aires. Otro punto estrategico desaprovechado es Pompeya donde a unas pocas cuadras tenemos el subte y el tren Belgrano Sur, luego de unas cuadras ya estas en Provincia de Buenos Aires.

**El sesgo hacia el norte y la deuda con el sur**

Es importante si queremos superar el hito historico que se consiguio en 2019 con mas de 6 millones de viajes incorporar nuevos puntos estrategicos para que las miles de personas que desde la provincia de Buenos Aires vienen todos los dias hacia capital para trabajar, para estudiar y para pasear puedan acceder y recorrer rapidamente la capital porteña del tango y el dulce de leche con las ecobicis.

Reforzar lo que ya funciona, como poner nuevas ciclos vias dentro de la misma localidades no aumenta la cantidad de viajes.

Recomendamos exteneder y llevas las ciclovias hacia nuevos publicos, de generación Milennials jovenes entre 25-34 años.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargamos tu tabla final de diagnóstico
datos_dias = {
    'Tipo_de_Dia': ['1. Día Hábil (Lun-Vie)', '2. Fin de Semana (Sáb/Dom)'],
    'Cantidad_Viajes': [2410010, 580548],
    'Promedio_Minutos': [25.99, 32.08]
}
df_dias = pd.DataFrame(datos_dias)

# 2. Creamos un lienzo dividido en 2 columnas (1 fila, 2 columnas)
plt.style.use('seaborn-v0_8-darkgrid')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 3. Primer Gráfico: El Volumen de Viajes (Azul)
sns.barplot(x='Tipo_de_Dia', y='Cantidad_Viajes', data=df_dias, palette='Blues_r', ax=ax1)
ax1.set_title('Volumen: Días Hábiles vs. Fines de Semana', fontsize=14, pad=10)
ax1.set_xlabel('Tipo de Día', fontsize=12)
ax1.set_ylabel('Cantidad de Viajes (en millones)', fontsize=12)

# 4. Segundo Gráfico: El Tiempo de Bienestar (Verde)
sns.barplot(x='Tipo_de_Dia', y='Promedio_Minutos', data=df_dias, palette='Greens_r', ax=ax2)
ax2.set_title('Bienestar: Duración Promedio del Viaje', fontsize=14, pad=10)
ax2.set_xlabel('Tipo de Día', fontsize=12)
ax2.set_ylabel('Minutos Promedio', fontsize=12)

# 5. Mostramos la obra de arte
plt.tight_layout()
plt.show()
```

De lunes a viernes tenemos un volumen masivo de 2.41 nillones de viajes de una duración corta.

Las personas usan la bicicleta con un proposito estricto: ir a trabajar, conectar con el tren, llegar a la facultad. El objetivo no es pasear sino llegar a otro destino rapido y economico.

Los fines de semana  el volumen total de viaje cae a 580 mil viajes, son dos dias y no son laborales. El tiempo de ocio incrementa los minutos de pedaleo.
