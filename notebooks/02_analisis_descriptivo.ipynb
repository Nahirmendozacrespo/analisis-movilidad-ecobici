```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargamos tu tabla anual
datos_anio = {
    'Anio': [2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025],
    'Total_Viajes': [2619968, 6217004, 2415597, 2860091, 2922805, 2622331, 3559284, 3156182]
}
df_anio = pd.DataFrame(datos_anio)

# 2. Configuramos el lienzo
plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots(figsize=(10, 5))

# 3. Dibujamos la línea evolutiva
sns.lineplot(
    x='Anio',
    y='Total_Viajes',
    data=df_anio,
    color='#3498db',
    marker='o',
    linewidth=3,
    markersize=10,
    ax=ax
)

# 4. Rellenamos el área bajo la curva para darle un toque "Senior"
ax.fill_between(df_anio['Anio'], df_anio['Total_Viajes'], color='#3498db', alpha=0.2)

# 5. Títulos y etiquetas
plt.title('Evolución Histórica: El Impacto de la Pandemia en EcoBici', fontsize=16, pad=15)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Cantidad de Viajes Totales (en millones)', fontsize=12)

# Mostramos el gráfico
plt.tight_layout()
plt.show()
```

Hasta el 2021 las ecobicis fueron de uso gratuito y a partir de Marzo de 2021 se mepzo a cobrar.

Las ecobicis comenzaron en 2010, luego en 2019 paso a manos de Timbici, mismo año que tuvieron la mayor cantidad de viajes y ya para el 2020 fue la pandemia del Covid-19.

Los siguientes años se mantuvieron con una pequeña subida en viajes totales. Podemos ver en el grafico una subida continua.
