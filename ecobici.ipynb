```python
import pandas as pd
import glob
```

**Concatenado bloque 2**

```python
# 1. Define la ruta a la carpeta donde están TODOS tus CSVs.
# El "/*.csv" al final es clave: le dice a Python que busque todos los archivos con esa extensión.
# IMPORTANTE: Reemplaza TU_CARPETA con la ruta real.
ruta_carpeta = '/content/drive/MyDrive/ecobici/recorridos/*.csv'

# 2. Encontrar todos los archivos CSV en esa carpeta
archivos = glob.glob(ruta_carpeta)
print(f"Se encontraron {len(archivos)} archivos para unir.")

# 3. Crear una lista vacía para guardar temporalmente cada año
lista_dataframes = []

# 4. Leer cada archivo uno por uno y agregarlo a la lista
for archivo in archivos:
    # Leemos el archivo
    df_temporal = pd.read_csv(archivo)
    # Lo agregamos a la lista
    lista_dataframes.append(df_temporal)

# 5. Unir (concatenar) todos los años en un solo "súper" DataFrame
# ignore_index=True asegura que las filas se enumeren del 0 hasta el final sin repetir números
df_historico = pd.concat(lista_dataframes, ignore_index=True)

# 6. Ver el resultado final
print("Tamaño del dataset histórico (Filas, Columnas):")
print(df_historico.shape)
```

Verificamos que los datataset desde el 2015 al 2017 tiene menos columnas que los dataset del 2023 al 2025.



*   2015-2017 no tienen id_usuario y id_recorridos. No es posible usarlos como llave para unirlos en SQL.
*   2023-2025 tienen las mismas columnas con diferente formato. Por ejemplo: id_usuario en el dataset usuarios esta como int64 mientras que id_usuario en el dataset recorridos esta como object, tiene un sufijo luego del numero "EcoBici"

Vamos a borrar el sufijo de id_usuario para dejar solo los numeros y luego corroborar si son datos unicos que coinciden en ambos dataset.

Que sean datos unicos nos serviria para unir las tablas en SQL.



```python
# 1. Estandarizar el nombre de la columna para que se llamen EXACTAMENTE igual
df_usuarios2023 = df_usuarios2023.rename(columns={'ID_usuario': 'id_usuario'})
```

```python
# 2. Convertir ambas a texto (nuestra regla de oro)
df_usuarios2023['id_usuario'] = df_usuarios2023['id_usuario'].astype(str)
df_historico2023['id_usuario'] = df_historico2023['id_usuario'].astype(str)
```

```python
# 3. Limpiar la "basura" de la tabla de recorridos
df_historico2023['id_usuario'] = df_historico2023['id_usuario'].str.replace('BAEcobici', '', regex=False)
```

```python
# --- LA PRUEBA DE FUEGO DE LA LLAVE ---

# A. Verificamos que sea una "Llave Primaria" válida (sin duplicados)
usuarios_duplicados = df_usuarios2023['id_usuario'].duplicated().sum()
print(f"Usuarios duplicados en la tabla base: {usuarios_duplicados}")
if usuarios_duplicados == 0:
    print("¡Excelente! La columna funciona como una Llave Primaria perfecta.")
else:
    print("Cuidado: Hay IDs repetidos en la tabla de usuarios, habrá que limpiarlos.")

print("\n" + "-"*40 + "\n")

# B. El Momento de la Verdad: ¿Qué porcentaje de viajes cruzará con éxito?
viajes_con_match = df_historico2023['id_usuario'].isin(df_usuarios2023['id_usuario'])
porcentaje_exito = (viajes_con_match.mean() * 100)

print(f"Total de viajes en 2023: {len(df_historico2023):,}")
print(f"Viajes que SÍ encontraron a su usuario: {viajes_con_match.sum():,}")
print(f"Tasa de éxito del cruce (JOIN): {porcentaje_exito:.2f}%")
```

Encontramos que hay recorridos sin id_usuario.

Vamos a crear una tabla con todos los usuarios desde 2015 hasta 2025.
Buscamos que todos los recorridos tengan su id_usuario.

```python
# 1. Asegurarnos de que ambas columnas sean texto (nuestra regla de oro para los IDs)
df_usuarios2023['id_usuario'] = df_usuarios2023['id_usuario'].astype(str)
df_historico2023['id_usuario'] = df_historico2023['id_usuario'].astype(str)

# 2. Extraer los usuarios únicos de cada tabla convirtiéndolos en "conjuntos" (sets)
usuarios_registrados = set(df_usuarios2023['id_usuario'])
usuarios_con_viajes = set(df_historico2023['id_usuario'])

# 3. Buscar la intersección (los IDs que están en ambas tablas a la vez)
coincidencias = usuarios_registrados.intersection(usuarios_con_viajes)

# Mostrar los resultados del cruce
print("--- RADIOGRAFÍA DE USUARIOS 2023 ---")
print(f"1. Total de IDs únicos en tabla Usuarios: {len(usuarios_registrados):,}")
print(f"2. Total de IDs únicos en tabla Recorridos: {len(usuarios_con_viajes):,}")
print(f"3. IDs que coinciden perfectamente en ambas: {len(coincidencias):,}")
print("-" * 36)

# 4. Ver el impacto real en los viajes con la función isin()
viajes_con_match = df_historico2023['id_usuario'].isin(usuarios_registrados)

print("\n--- IMPACTO EN LOS VIAJES ---")
print("¿Cuántos viajes tienen un id_usuario que SÍ existe en la tabla Usuarios?")
print(viajes_con_match.value_counts())
```
