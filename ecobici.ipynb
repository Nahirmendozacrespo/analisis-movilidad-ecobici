**Concatenado bloque 2**

```python
# 1. Define la ruta a la carpeta donde están TODOS tus CSVs.
# El "/*.csv" al final es clave: le dice a Python que busque todos los archivos con esa extensión.
# IMPORTANTE: Reemplaza TU_CARPETA con la ruta real.
ruta_carpeta = '/content/drive/MyDrive/ecobici/recorridos/*.csv'

# 2. Encontrar todos los archivos CSV en esa carpeta
archivos = glob.glob(ruta_carpeta)
print(f"Se encontraron {len(archivos)} archivos para unir.")

# 3. Crear una lista vacía para guardar temporalmente cada año
lista_dataframes = []

# 4. Leer cada archivo uno por uno y agregarlo a la lista
for archivo in archivos:
    # Leemos el archivo
    df_temporal = pd.read_csv(archivo)
    # Lo agregamos a la lista
    lista_dataframes.append(df_temporal)

# 5. Unir (concatenar) todos los años en un solo "súper" DataFrame
# ignore_index=True asegura que las filas se enumeren del 0 hasta el final sin repetir números
df_historico = pd.concat(lista_dataframes, ignore_index=True)

# 6. Ver el resultado final
print("Tamaño del dataset histórico (Filas, Columnas):")
print(df_historico.shape)
```

# **Tabla usuarios 2015-2025**

```python
import pandas as pd
import glob

# 1. Definir la ruta específica para buscar solo los archivos de usuarios
# El asterisco (*) atrapará cualquier año en el nombre del archivo
# Asegúrate de que la ruta incluya tu carpeta 'ecobici' y el comodín '*.csv' al final
ruta_usuarios = '/content/drive/MyDrive/ecobici/usuarios-ecobici-*.csv'
archivos_usuarios = glob.glob(ruta_usuarios)
print(f"Se encontraron {len(archivos_usuarios)} archivos de usuarios para unir.")

lista_usuarios = []

for archivo in archivos_usuarios:
    df_temp = pd.read_csv(archivo)

    # 2. Estandarizar el nombre de la columna (por el problema de ID_usuario vs id_usuario)
    if 'ID_usuario' in df_temp.columns:
        df_temp = df_temp.rename(columns={'ID_usuario': 'id_usuario'})

    # 3. Forzar el ID a texto y limpiarlo de decimales (.0) o espacios
    if 'id_usuario' in df_temp.columns:
        df_temp['id_usuario'] = df_temp['id_usuario'].astype(str)
        df_temp['id_usuario'] = df_temp['id_usuario'].str.replace('.0', '', regex=False).str.strip()

    lista_usuarios.append(df_temp)

# 4. Concatenar toda la historia en una sola tabla gigante
df_usuarios_historico = pd.concat(lista_usuarios, ignore_index=True)
total_previo = len(df_usuarios_historico)

# 5. LA PRUEBA DE FUEGO: Eliminar usuarios duplicados
# subset=['id_usuario'] le dice a Pandas que busque repetidos basándose solo en el ID
# keep='last' asegura que si hay duplicados, nos quedamos con la versión más reciente de sus datos
df_usuarios_historico = df_usuarios_historico.drop_duplicates(subset=['id_usuario'], keep='last')
total_final = len(df_usuarios_historico)

print("\n--- RESUMEN DEL PADRÓN MAESTRO ---")
print(f"Registros brutos concatenados: {total_previo:,}")
print(f"Duplicados eliminados: {total_previo - total_final:,}")
print(f"🚴 Usuarios ÚNICOS finales: {total_final:,}")
```

```python
# 2. Convertir ambas a texto (nuestra regla de oro)
df_usuarios_historico['id_usuario'] = df_usuarios_historico['id_usuario'].astype(str)
df_historico2023['id_usuario'] = df_historico2023['id_usuario'].astype(str)
```

```python
# 3. Limpiar la "basura" de la tabla de recorridos
df_historico2023['id_usuario'] = df_historico2023['id_usuario'].str.replace('BAEcobici', '', regex=False)
```

```python
# 1. El Momento de la Verdad: ¿Qué porcentaje de viajes cruza con el Padrón Maestro?
# Usamos isin() para buscar cada ID de recorrido dentro del historial completo de usuarios
viajes_con_match_historico = df_historico2023['id_usuario'].isin(df_usuarios_historico['id_usuario'])

# 2. Calcular el porcentaje de éxito
porcentaje_exito_historico = (viajes_con_match_historico.mean() * 100)

# 3. Mostrar los resultados triunfales
print("--- PRUEBA DE FUEGO CON EL PADRÓN MAESTRO ---")
print(f"Total de viajes en 2023: {len(df_historico2023):,}")
print(f"Viajes que SÍ encontraron a su usuario: {viajes_con_match_historico.sum():,}")
print(f"Tasa de éxito del cruce (JOIN) actualizada: {porcentaje_exito_historico:.2f}%")
```

Tenemos llave para unir las tablas en SQL con ID usuario entre tablas recorridos y usuarios. Tambien tenemos Padron historico del 2015-2025.

El cruce del padron con la tabla del 2023 "id_usuario" dio un 94.34% de confiabilidad.

Limpieza de columnas administrativas

```python
# 1. Definimos una lista con los nombres EXACTOS de las columnas a eliminar
columnas_basura = [
    'ID Cliente',
    'Customer Has Dni (Yes / No)',
    'Customer.Has.Dni..Yes...No.'
]

# 2. Le pedimos a Pandas que las elimine (drop) del Padrón Maestro
df_usuarios_historico = df_usuarios_historico.drop(columns=columnas_basura)

# 3. Volvemos a correr un info() rápido para confirmar que desaparecieron
print("--- RADIOGRAFÍA FINAL DEL PADRÓN LIMPIO ---")
df_usuarios_historico.info()
```

vamos a cambiar el tipo de datos de la columa fecha

```python
# Convertimos el texto a Fecha Real permitiendo formatos mezclados (mixed)
# dayfirst=True le da la pista a Pandas de que, en las fechas cortas, el primer número es el día
df_usuarios_historico['fecha_alta'] = pd.to_datetime(
    df_usuarios_historico['fecha_alta'],
    format='mixed',
    dayfirst=True
)

# Verificamos que ahora sí funcionó
print("Tipos de datos actualizados:")
print(df_usuarios_historico.dtypes)
print("\nPrimeras fechas modernizadas:")
print(df_usuarios_historico[['fecha_alta']].head())
```

cambiamos tipo de dato de edad a float64

```python
# 1. Forzar la columna de edad a ser estrictamente numérica
df_usuarios_historico['edad_usuario'] = pd.to_numeric(df_usuarios_historico['edad_usuario'], errors='coerce')

# 2. Confirmar el cambio y ver estadísticas de la edad (promedio, mínimo y máximo)
print(df_usuarios_historico['edad_usuario'].describe())
```

```python
import numpy as np
```

```python
# 1. Definimos los límites de lo que consideramos una "edad humana lógica"
edad_minima = 16
edad_maxima = 100

# 2. Reemplazamos las edades fuera de este rango por NaN (Nulo)
df_usuarios_historico.loc[
    (df_usuarios_historico['edad_usuario'] < edad_minima) |
    (df_usuarios_historico['edad_usuario'] > edad_maxima),
    'edad_usuario'
] = np.nan

# 3. Volvemos a mirar las estadísticas para confirmar la limpieza
print("--- ESTADÍSTICAS DE EDAD LIMPIAS ---")
print(df_usuarios_historico['edad_usuario'].describe())
```

Exportamos padron de usuarios del 2015-2025
