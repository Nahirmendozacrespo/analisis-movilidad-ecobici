# **Expansión inteligente y segura de las EcoBicis**

**Contexto**

En nuestros analisis demostramos que las ecobicis se utilizan en horas claves para el traslado de un destino a otro en horario de trabajo, universidad y actidades puntuales.

La expansión de las ciclovias y puntos de ecobicis se repiten en las mismas zonas donde ya estan. Esto se traduce en que los ultimos años los viajes se mantienen pero el crecimiento es lento. Reforzar las Ecobicis en los mismos puntos no llega a mas usuarios y no genera mas recorridos.

Identificamos un sesgo hacia el norte y el centro de la ciudad.

Las zonas estrategicas de trasnferencia con la provincia de Buenos Aires en el sur (como la conexión con el ferrocarril Belgrano Sur) estan aisladas. La expansión a zonas vulnerables se ve frenada por el riesgo de perdida, vandalismo y robo de unidades, lo que aleja a los patrocinadores.

**Problema:**

Las estaciones mas importantes como Retiro, Constitución y Pacífico son puntos claves y estrategicos del transporte urbano donde se encuentran trenes, subtes, colectivos y bicicletas al mismo tiempo.

Se genera trafico, demoras, riesgos de accidentes.

**Solución**

Pnesemos en niveles de circulación.

🚲 Bici-tuneles
Tuneles exlusivos para bicicletas

🚲 Bici-puentes
Puentes exclusivos para bicicletas y peatones por arriba de las zonas complecada.

**Ventajas**

* Circulación continua
* Cruces más rápidos
* Seguridad total
* Menor impacto visual

**El plan de acción**

Democratizar el uso de la bicicleta hacia el sur garatanzando un retorno de inversión del patrocinador y seguridad de los activos, el gobierno de la ciudad y tembici no deben dispersar bicicletas sueltas sino implementar un modelo de "Nodos de transferencia segura"

1. Infraestructura de Zonas Cuidadas

* En lugar de publicidad tradicional, Banco Macro puede financiar Estaciones Blindadas, ubicadas en la boca de los trenes.

Estas estaciones funcionan como zonas seguras: microespacios bien iluminados y protegidos por la policia federal dentro de la boca de los trenes.

2. Validación de identidad: La extracción de bicicletas debe requerir una validación robusta vinculada a una tarjeta de credito, app bancaria. Cruzar al usuario con un medio de pago real.

3. Corredores seguros exclusivos: las zonas estrategicas son en esencia zonas de enbotellamiento. La red de ciclovias que conecte el centro porteño con el sur debe ser exclusivo para bicicletas que lleven al usuario de la boca del tren hacia la boca del subte u otro punto estrategico como puede ser once o constitución.


**Como financiarlo**

Estas obras podrian financiarse con inversión publica y tambien privada.

Por ejemplo: *"Tunel sustentable Macro"*  *"Puente sustentable Macro"*

**Resumen**

Para que Ecobici crezca y supere su hito historico de viajes en 2019, Buenos Aires necesita modernizarse y crear infraestructura mas inteligente para llegar a millones de personas que hoy estan excluidas del servicio que promete accesibilidad.
